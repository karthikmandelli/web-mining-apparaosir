```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Text
text1 = """Guacamole Turkey Burgers instructions In a medium bowl, combine the turkey, onion, chili powder, lime zest, and half the cilantro. Season with the measured salt and some black pepper. Use your hands to gently combine. Divide the meat into 4 equal portions and shape each portion into an even 1/2-inch-thick patty. Use your thumb to make a shallow indentation in the center of each patty (this helps the burgers cook more evenly). Transfer the patties to a plate and set aside. To make the guacamole, place the avocado in a small bowl and, with the back of a fork, mash it until it’s almost smooth. Add the tomato, the remaining cilantro, and the lime juice. Season with salt and stir to combine. Heat a grill pan or outdoor grill to medium-high, about 375°F to 425°F. When the grill is hot, use tongs to grab several layers of wadded paper towels, dip them in vegetable oil, and smear the grill pan’s the top of each patty for the last 4 minutes of cooking time. When the cheese melts, transfer the burgers to a warmed plate and tent loosely with foil for 5 minutes. Meanwhile, split and toast the buns on the grill. Top each bottom bun with a turkey patty, divide the guacamole evenly among the patties, and cover with the bun tops. Serve immediately."""

# Tokenization
tokens = word_tokenize(text1)

# Remove stopwords
stopwords = set(stopwords.words('english'))
tokens_without_stopwords = [token for token in tokens if token.lower() not in stopwords]

# Prepare keywords/Index words
keywords = list(set(tokens_without_stopwords))

# Print the results
print("Keywords/Index Words:")
for keyword in keywords:
    print(keyword)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tokens = text1.split()
processed_text = ' '.join(tokens)
# Calculate TF-IDF
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform([processed_text])
# Get the feature names (terms)
feature_names = tfidf_vectorizer.get_feature_names_out()
# Print the TF-IDF scores
for i, term in enumerate(feature_names):
    tfidf_score = tfidf_matrix[0, i]
    print(f"Term: {term}, TF-IDF Score: {tfidf_score}")
```

#**Calculate term weights using TF-IDF**

##we are calculating the TF-IDF by removing the stopwords

```python
import zipfile
import nltk
from nltk.corpus import stopwords
import string
from sklearn.feature_extraction.text import TfidfVectorizer

# Download the 'punkt' resource and stopwords
nltk.download('punkt')
nltk.download('stopwords')

# Path to the ZIP file
zip_file_path = "/content/archive.zip"

# Name of the folder you want to extract and tokenize
folder_name = "technologie"

# Get the list of stopwords and punctuation
stopwords = set(stopwords.words('english'))
punctuation = set(string.punctuation)

# Open the ZIP file
with zipfile.ZipFile(zip_file_path, 'r') as zip_file:
    # Get a list of all files and folders in the ZIP
    file_list = zip_file.namelist()

    # Iterate over each file/folder
    for file_name in file_list:
        # Check if the file/folder is in the desired folder
        if file_name.startswith(folder_name + "/"):
            # Extract and tokenize the content of the file
            with zip_file.open(file_name) as file:
                content = file.read().decode("utf-8")
                tokens = nltk.word_tokenize(content)

                # Remove stopwords and punctuation
                tokens = [token for token in tokens if token.lower() not in stopwords and token not in punctuation]

                # Calculate TF-IDF
                vectorizer = TfidfVectorizer()
                tfidf_matrix = vectorizer.fit_transform([" ".join(tokens)])

                # Get the feature names (terms)
                feature_names = vectorizer.get_feature_names_out()

                # Print the term weights (TF-IDF values)
                for i, term in enumerate(feature_names):
                    weight = tfidf_matrix[0, i]
                    print(f"{term}: {weight}")
```

##TF-IDF is high  means it will shows the high importanace in the document or uniqueness of the term

```python
import zipfile
import nltk
from nltk.corpus import stopwords
import string
from sklearn.feature_extraction.text import TfidfVectorizer

# Download the 'punkt' resource and stopwords
nltk.download('punkt')
nltk.download('stopwords')

# Path to the ZIP file
zip_file_path = "/content/archive.zip"

# Name of the folder you want to extract and tokenize
folder_name = "technologie"

# Get the list of stopwords and punctuation
stopwords = set(stopwords.words('english'))
punctuation = set(string.punctuation)

# Open the ZIP file
with zipfile.ZipFile(zip_file_path, 'r') as zip_file:
    # Get a list of all files and folders in the ZIP
    file_list = zip_file.namelist()

    # Iterate over each file/folder
    for file_name in file_list:
        # Check if the file/folder is in the desired folder
        if file_name.startswith(folder_name + "/"):
            # Extract and tokenize the content of the file
            with zip_file.open(file_name) as file:
                content = file.read().decode("utf-8")
                tokens = nltk.word_tokenize(content)

                # Remove stopwords and punctuation
                tokens_without_stopwords = [token for token in tokens if token.lower() not in stopwords and token not in punctuation]

                # Calculate TF-IDF with stopwords
                vectorizer_with_stopwords = TfidfVectorizer()
                tfidf_matrix_with_stopwords = vectorizer_with_stopwords.fit_transform([" ".join(tokens)])

                # Calculate TF-IDF without stopwords
                vectorizer_without_stopwords = TfidfVectorizer()
                tfidf_matrix_without_stopwords = vectorizer_without_stopwords.fit_transform([" ".join(tokens_without_stopwords)])

                # Get the feature names (terms) for both versions as Python lists
                feature_names_with_stopwords = vectorizer_with_stopwords.get_feature_names_out().tolist()
                feature_names_without_stopwords = vectorizer_without_stopwords.get_feature_names_out().tolist()

                # Print terms present in both versions and their difference in TF-IDF weights
                for term in set(feature_names_with_stopwords).intersection(feature_names_without_stopwords):
                    index_with_stopwords = feature_names_with_stopwords.index(term)
                    index_without_stopwords = feature_names_without_stopwords.index(term)

                    weight_with_stopwords = tfidf_matrix_with_stopwords[0, index_with_stopwords]
                    weight_without_stopwords = tfidf_matrix_without_stopwords[0, index_without_stopwords]

                    difference = abs(weight_with_stopwords - weight_without_stopwords)

                    print(f"{term}: Difference = {difference}")
```

To rank documents using the Boolean model and Vector space model, you need to follow these steps:

Preprocessing:

Tokenize the query and documents: Split the query and document sentences into individual terms or tokens.
Apply stemming or lemmatization: Normalize the terms to their base or root form to handle variations.
Boolean Model:

Convert the query and documents into Boolean representations: Create Boolean vectors indicating the presence or absence of each term in the query and documents.
Perform Boolean operations: Apply AND, OR, and NOT operations between the query and document vectors to determine the matching documents.
Rank the documents: The documents that match the query based on Boolean operations are considered relevant.
Vector Space Model:

Build the term-document matrix: Create a matrix where each row represents a document and each column represents a term, and each cell contains the weight (e.g., TF-IDF) of the term in the corresponding document.
Calculate query-document similarity: Compute the similarity between the query vector and each document vector using techniques like cosine similarity.
Rank the documents: Sort the documents based on their similarity scores in descending order. The higher the similarity, the more relevant the document.
Note that the implementation details and specific algorithms for each step may vary depending on the programming language and libraries you use. Libraries like scikit-learn or nltk in Python provide useful tools for vectorization and similarity calculations.

Keep in mind that the Vector space model is generally more flexible and effective for ranking documents compared to the Boolean model, as it considers the weight and importance of terms. The Boolean model, on the other hand, simply evaluates the presence or absence of terms without considering their relevance or importance.

**Boolean Model:**

Create a boolean vector representation of the query.

Initialize a boolean vector with all values set to False.

For each term in the query, set the corresponding position in the boolean vector to True.

The boolean vector represents the presence or absence of each term in the query.

**Vector Space Model:**

Create a vector space representation of the query.

Initialize a vector with all values set to 0.

For each term in the query, calculate its term frequency (TF) in the query.

If the term is present in the vocabulary of the corpus, retrieve its inverse document frequency (IDF).

Multiply the TF of the term by its IDF and assign the result to the corresponding position in the vector.

The vector represents the importance or weight of each term in the query.

```python
import nltk
from nltk.corpus import stopwords
import string
from sklearn.feature_extraction.text import TfidfVectorizer

# Download the 'punkt' resource and stopwords
nltk.download('punkt')
nltk.download('stopwords')

# Define the query
query = "Why worry about mobile phones. Soon they will be subsumed into the PDA's / laptops etc."

# Preprocessing
tokens = nltk.word_tokenize(query)
stopwords = set(stopwords.words('english'))
punctuation = set(string.punctuation)
tokens = [token.lower() for token in tokens if token.lower() not in stopwords and token not in punctuation]

# Boolean Model
boolean_vector = [False] * len(tokens)
for i, term in enumerate(tokens):
    boolean_vector[i] = True

# Vector Space Model
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([" ".join(tokens)])
feature_names = vectorizer.get_feature_names_out()
vector_space_vector = tfidf_matrix.toarray()[0]

# Print the boolean vector
print("Boolean Vector:")
print(boolean_vector)

# Print the vector space vector
print("\nVector Space Vector:")
for i, term in enumerate(feature_names):
    weight = vector_space_vector[i]
    print(f"{term}: Weight = {weight}")
```
